#### Question 1.3
Formulate an hypothesis about pain severity before and after treatment in 24 weeks.

```python
data = data.loc[:,['FIQ_baseline','FIQ_12W','FIQ_24W']]
```

H0: The null hypothesis is that the means of all three groups is the same - There is no difference in pain level before and after the treatment
H1: The means of all three groups are different.

```python
GrandMean = (data['FIQ_baseline'].sum()+ data['FIQ_12W'].sum()+data['FIQ_24W'].sum())/150
print( "FIQ_baseline mean: %s  FIQ_12W mean: %s FIQ_24W mean: %s and the GrandMean: %s" %(data['FIQ_baseline'].mean(),  data['FIQ_12W'].mean(), data['FIQ_24W'].mean(),GrandMean ))
```

```python
SST  = ((data['FIQ_baseline'] - GrandMean)**2).sum()
SSTG = 50*((data['FIQ_baseline'].mean()- GrandMean)**2).sum() + 50*((data['FIQ_12W'].mean()- GrandMean)**2).sum() + 50*((data['FIQ_24W'].mean()- GrandMean)**2).sum()
SSTE = ((data['FIQ_baseline'] - data['FIQ_baseline'].mean())**2).sum() + ((data['FIQ_12W'] - data['FIQ_12W'].mean())**2).sum() + ((data['FIQ_24W'] - data['FIQ_24W'].mean())**2).sum()


#To make the sum of squares comparable, we divide each one by their associated degrees of freedom
MSTG = SSTG/(3-1)
MSTE = SSTE/(50-3)

#The test statistic is the ratio of group and error mean squares
F = MSTG/MSTE
```

```python
print("The p-value =",1-stats.f.cdf(F, dfn=2, dfd=12))
```

#### Question 1.4 
Discuss your results.

Since the p-value is more than the assumed significance value of 0.05, there is no significant evidence against the null hypothesis.

H0: The null hypothesis is that the means of all three groups is the same
H1: The means of all three groups are different.

```python
GrandMean = (data['A'].sum()+ data['B'].sum()+data['C'].sum())/45
print( "A mean: %s  B mean: %s C mean: %s and the GrandMean: %s" %(data['A'].mean(),  data['B'].mean(), data['C'].mean(),GrandMean ))
```

```python
SST  = ((data['A'] - GrandMean)**2).sum()
SSTG = 15*((data['A'].mean()- GrandMean)**2).sum() + 15*((data['B'].mean()- GrandMean)**2).sum() + 15*((data['C'].mean()- GrandMean)**2).sum()
SSTE = ((data['A'] - data['A'].mean())**2).sum() + ((data['B'] - data['B'].mean())**2).sum() + ((data['C'] - data['C'].mean())**2).sum()


#To make the sum of squares comparable, we divide each one by their associated degrees of freedom
MSTG = SSTG/(3-1)
MSTE = SSTE/(15-3)

#The test statistic is the ratio of group and error mean squares
F = MSTG/MSTE
```

```python
print("The p-value =",1-stats.f.cdf(F, dfn=2, dfd=12))
```

Since the p-value is less that the assumed significance value of 0.05, there is significant evidence against the null hypothesis.
